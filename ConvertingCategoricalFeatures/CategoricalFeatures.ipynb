# Converting Categorical Features

### Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Get the Data

```python
titanic = pd.read_csv('D:\\P Y T H O N\\Machine Learning\\ConvertingCategoricalFeatures\\titanic.csv')
titanic.head()
```

```python
titanic['Sex'].unique()
```

```python
Sex = pd.get_dummies(titanic['Sex'])
Sex
```

```python
Sex = pd.get_dummies(titanic['Sex'], drop_first=True)
Sex
```

```python
titanic['Embarked'].unique()
```

```python
titanic.dropna(inplace=True)
```

```python
titanic['Embarked'].unique()
```

```python
Embarkes = pd.get_dummies(titanic['Embarked'], drop_first=True)
Embarkes
```

```python
titanic.head()
```

```python
titanic.drop(['PassengerId','Name','Ticket','Sex','Embarked'], axis=1, inplace=True)
titanic.head()
```

```python
titanic = pd.concat([titanic, Sex, Embarkes], axis=1)
titanic.head()
```
